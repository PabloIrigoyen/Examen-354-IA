```python
import pandas as pd
data = pd.read_csv('rpg_steam_clean.csv')
datos = data['release_price'].dropna().tolist()
datos = [float(x) for x in datos]
def normalizacion(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]
def l1(coef):
    return sum(abs(c) for c in coef)

def l2(coef):
    return sum(c ** 2 for c in coef)
normal = normalizacion(datos)
pl1 = l1(normal)
pl2 = l2(normal)
print(f"Datos normalizados: {normal}")
print(f"Penalización L1: {pl1}")
print(f"Penalización L2: {pl2}")
```
