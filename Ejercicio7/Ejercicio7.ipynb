```python
import random
from deap import base, creator, tools, algorithms
datos=[5, 4, 1, 3, 8, 2, 7, 6, 5, 4, 3, 1]
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)
toolbox = base.Toolbox()
toolbox.register("individual", lambda: creator.Individual(random.sample(datos, len(datos))))
toolbox.register("population", tools.initRepeat, list, toolbox.individual, n=12)
def evaluate(individual):
    return sum(individual),
toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=3)
population = toolbox.population()
NGEN=3
CXPB=0.5
MUTPB=0.2
for gen in range(NGEN):
    offspring = toolbox.select(population, len(population))
    offspring = list(map(toolbox.clone, offspring))
    for child1, child2 in zip(offspring[::2], offspring[1::2]):
        if random.random() < CXPB:
            toolbox.mate(child1, child2)
            del child1.fitness.values
            del child2.fitness.values
    for mutant in offspring:
        if random.random() < MUTPB:
            toolbox.mutate(mutant)
            del mutant.fitness.values
    invalid_ind = [ind for ind in offspring if not ind.fitness.valid]
    fitnesses = map(toolbox.evaluate, invalid_ind)
    for ind, fit in zip(invalid_ind, fitnesses):
        ind.fitness.values = fit
    population[:] = offspring
    print(f"-- Generación {gen + 1}")
    for ind in population:
        print(ind)
best_ind = tools.selBest(population, 1)[0]
print(f"\nMejor individuo: {best_ind}, Fitness: {best_ind.fitness.values[0]}")
```

```python
import random
datos=[5, 4, 1, 3, 8, 2, 7, 6, 5, 4, 3, 1]
TAMANO_POBLACION=12
TAMANO_INDIVIDUO=len(datos)
NGEN=3
CXPB=0.5
MUTPB=0.2
def inicializar_poblacion(tamano_poblacion):
    return [random.sample(datos, TAMANO_INDIVIDUO) for _ in range(tamano_poblacion)]
def evaluar(individuo):
    return sum(individuo)
def seleccion_torneo(poblacion, k=3):
    return max(random.sample(poblacion, k), key=evaluar)
def cruzar(padre1, padre2):
    punto1,punto2=sorted(random.sample(range(TAMANO_INDIVIDUO), 2))
    hijo1=padre1[:punto1]+padre2[punto1:punto2]+padre1[punto2:]
    hijo2=padre2[:punto1]+padre1[punto1:punto2]+padre2[punto2:]
    return hijo1, hijo2
def mutar(individuo):
    punto=random.randint(0,TAMANO_INDIVIDUO-1)
    nuevo_valor=random.choice(datos)
    individuo[punto]=nuevo_valor
    return individuo
poblacion = inicializar_poblacion(TAMANO_POBLACION)
for gen in range(NGEN):
    nueva_poblacion = []
    while len(nueva_poblacion) < TAMANO_POBLACION:
        padre1=seleccion_torneo(poblacion)
        padre2=seleccion_torneo(poblacion)
        if random.random()<CXPB:
            hijo1,hijo2=cruzar(padre1, padre2)
        else:
            hijo1,hijo2=padre1[:], padre2[:]
        if random.random()<MUTPB:
            hijo1=mutar(hijo1)
        if random.random()<MUTPB:
            hijo2 = mutar(hijo2)
        nueva_poblacion.extend([hijo1, hijo2])
    poblacion=nueva_poblacion[:TAMANO_POBLACION]
    print(f"-- Generación {gen + 1} --")
    for ind in poblacion:
        print(ind)
mejor_individuo = max(poblacion, key=evaluar)
print(f"\nMejor individuo: {mejor_individuo}, Fitness: {evaluar(mejor_individuo)}")
```
