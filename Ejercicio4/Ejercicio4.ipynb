```python
#Label Encoding
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv('rpg_steam_clean.csv')
label_encoder = LabelEncoder()
df['release_price'] = label_encoder.fit_transform(df['release_price'])
print("\nDatos con etiqueta (Label Encoding):")
print(df['release_price'])
```

```python
#One-Hot Encoding
import pandas as pd
df=pd.read_csv('rpg_steam_clean.csv')
df_encoded = pd.get_dummies(df['release_price'], columns=['release_price'])
df_encoded = df_encoded.astype(int)
print("\nDatos con One-Hot Encoding:")
print(df_encoded)
```

```python
#Normalizacion
from sklearn.preprocessing import MinMaxScaler
data=pd.read_csv('rpg_steam_clean.csv')
columna=data[['release_price']]
scaler=MinMaxScaler()
scaler_data=scaler.fit_transform(columna)
print(scaler_data)
```

```python
#Estandarizacion
from sklearn.preprocessing import StandardScaler
data=pd.read_csv('rpg_steam_clean.csv')
columna=data[['release_price']]
scaler=StandardScaler()
scaler_data=scaler.fit_transform(columna)
print(scaler_data)
```
